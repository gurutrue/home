```python
import os
from PIL import Image
from pillow_heif import register_heif_opener

# Enable HEIC support
register_heif_opener()

# Folder with your images
image_folder = r"C:\Users\rynsmns\Pictures\Asurion Reference Photos"

def convert_images_to_png(folder):
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        name, ext = os.path.splitext(filename)
        ext = ext.lower()

        if ext == ".png":
            continue

        output_path = os.path.join(folder, name + ".png")

        try:
            image = Image.open(file_path)
            image.save(output_path, "PNG")
            print(f"Converted {filename} -> {name}.png")

            if ext == ".heic":
                os.remove(file_path)
                print(f"Deleted original HEIC: {filename}")

        except Exception as e:
            print(f"Failed to convert {filename}: {e}")

convert_images_to_png(image_folder)
```
